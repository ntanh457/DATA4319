```julia
using Plots
theme(:dark)


# Define a simple function to explore minimization on.
C(w) = (w - 4)^2 + 3

# Plot C
plot(C, 
    xlims = (-8, 15),
    xaxis = "w",
    ylims = (-5, 100),
    yaxis = "Cost C(w)",
    legend = false)
```

```julia
# Make initial guess 
w_0 = 12.0

println("w_0 = $w_0")
println("")

scatter!([(w_0, C(w_0))], 
        color = "yellow")
```

```julia
# Define the derivative of C(w)
∇C(w) = 2.0*w - 8.0

# Set your step length
α = .34

# Apply gradient descent 
w_1 = w_0 - α*∇C(w_0)


println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1))],
        color = "yellow")
```

```julia
# Apply gradient descent 
w_2 = w_1 - α*∇C(w_1)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "yellow")
```

```julia
# Apply gradient descent 
w_3 = w_2 - α*∇C(w_2)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")
println("w_3 = $w_3 and |∇C(w_3)| = ", abs(∇C(w_3)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3))],
        color = "yellow")
```

```julia
#Apply gradient descent 
w_4 = w_3 - α*∇C(w_3)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")
println("w_3 = $w_3 and |∇C(w_3)| = ", abs(∇C(w_3)))
println("")
println("w_4 = $w_4 and |∇C(w_4)| = ", abs(∇C(w_4)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3)), (w_4, C(w_4))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3)), (w_4, C(w_4))],
        color = "yellow")
```

```julia
function grad(x)
    return 2*x+5*cos(x)
end
    
function cost(x)
    return x^2+5*sin(x)
end
        
function myGD(eta,x0)
    x=float([x0])
    for i=1:100
        x_new=x[i]-eta*grad(x[i])
        if abs(grad(x_new))<=10^-3
            break
        end
        append!(x,x_new)
    end
        
    return x
end
```

```julia
x1 = myGD(.1, -5)
```

```julia
x2 = myGD(.1, 5)
```

```julia
# Define a simple function to explore minimization on.
C(x) = x^2+5*sin(x)
# Plot C
plot(C, 
    xlims = (-6, 6),
    xaxis = "w",
    ylims = (-5, 35),
    yaxis = "Cost C(w)",
    legend = false)
```

```julia
for i=1:min(length(x1),length(x2))
    if i==1
        scatter!([(x1[i], C(x1[i]))], 
        color = "yellow")

        scatter!([(x2[i], C(x2[i]))], 
        color = "green")
    else
        plot!([(x1[i-1], C(x1[i-1])), (x1[i], C(x1[i]))],
        color = "lightblue")
        scatter!([(x1[i], C(x1[i]))], 
        color = "yellow")
        plot!([(x2[i-1], C(x2[i-1])), (x2[i], C(x2[i]))],
        color = "lightblue")
        scatter!([(x2[i], C(x2[i]))], 
        color = "green")
    end
end
scatter!()
```
